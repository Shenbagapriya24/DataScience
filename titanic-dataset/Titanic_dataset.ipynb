```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/Shenbagapriya24/DataScience/main/titanic-dataset/train.csv")
```

```python
df.head()
```

```python
df_subset = df.drop(['PassengerId','SibSp','Parch','Ticket','Embarked'],axis = 1)
```

```python
df_subset.head()
```

```python
df_subset.isna().sum()
```

```python
df[df.Age.isnull() == True]
```

```python
df_subset.dropna(subset = ["Cabin","Age"],inplace = True)
```

```python
df_subset.groupby(['Sex','Pclass'])["Survived"].count()
```

```python
df_subset.pivot_table(index = 'Sex',columns = 'Pclass',values = 'Survived')
```

```python
pv_table = pd.pivot_table(df_subset, index=["Sex","Pclass"], aggfunc={"Survived":"mean","Age":"median"})
pv_table
```
